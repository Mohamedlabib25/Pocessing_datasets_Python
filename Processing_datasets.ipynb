```python
#step_0


import pandas as pd
import numpy as np


#loading data for Energy
Energy=pd.read_excel("E:\DM\Python\data\Energy Indicators.xls") 

#dropping the undesired rows and cols 
#renaming cols
Energy= Energy.drop(Energy.columns[[0,1]],axis=1)
Energy=Energy.drop(Energy.index[228:266],axis=0)
Energy=Energy.drop(Energy.index[0],axis=0)
Energy.rename(columns={"Unnamed: 2":"Country"},inplace=True)
Energy.rename(columns={"Renewable Electricity Production":"% Renewable"},inplace=True)

#replacing the null values
Energy["Energy Supply"].replace("...",np.NaN,inplace=True)
Energy["Energy Supply per capita"].replace("...",np.NaN,inplace=True)


#converting petajoule to gigajoules 
Energy["Energy Supply"]=Energy["Energy Supply"].apply(lambda x:x*1000000)


#Removing numbers and/or parenthesis in their name of country column
Energy['Country'] = Energy['Country'].str.replace('\d+', '')
Energy['Country']=Energy['Country'].str.replace(r"\(.*\)","")
Energy['Country']= Energy['Country'].str.strip()


#replacing names of some countries as required
Energy['Country']= Energy['Country'].replace({"Republic of Korea": "South Korea",
    "United States of America": "United States",
    "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
    "China, Hong Kong Special Administrative Region": "Hong Kong"})
display(Energy)
```

```python
#step_1


# loading data for GDP 
GDP=pd.read_csv("E:\DM\Python\data\world_bank.csv",header=4)

#Dropping undesirred rows
GDP=GDP.drop(GDP.index[0:4],axis=0)

#replacing the names of countries as required
GDP["Country Name"]=GDP["Country Name"].replace({"Korea, Rep.": "South Korea",  "Iran, Islamic Rep.": "Iran", "Hong Kong SAR, China": "Hong Kong"})

#getting only years from 2006 to 2015
GDP=GDP.drop(['Country Code', 'Indicator Name', 'Indicator Code',
       '1960', '1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968',
       '1969', '1970', '1971', '1972', '1973', '1974', '1975', '1976', '1977',
       '1978', '1979', '1980', '1981', '1982', '1983', '1984', '1985', '1986',
       '1987', '1988', '1989', '1990', '1991', '1992', '1993', '1994', '1995',
       '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004',
       '2005',], axis=1)
display(GDP)
```

```python
#step_2



#loading data for ScimEn
ScimEn_all=pd.read_excel("E:\DM\Python\data\scimagojr-3.xlsx")

#getting the top 15 countries by rank
ScimEn=ScimEn_all[0:15]
display(ScimEn)
```

```python
#step_3


#joing the ScimEn and Energy dataframes on the values of country col(it exists on both of them)
Sci_Energy=pd.merge(ScimEn, Energy, how='inner', on='Country')

#joing the Sci_Energy(the previous result) and gdp dataframes on the values of country col from the Sci_Energy side 
#and on the country name col from the gdp side
Sci_Energy_GDP=pd.merge(Sci_Energy, GDP, how='inner', left_on='Country', right_on="Country Name")

#making the country col as the index
Sci_Energy_GDP=Sci_Energy_GDP.drop(['Country Name'], axis=1)
Sci_Energy_GDP=Sci_Energy_GDP.set_index('Country')
display(Sci_Energy_GDP)
```

```python
#step_4


#joining the 3 dataframes agian but with the total rows on the ScimEn (ScimEn_all)
Sci_Energy_all=pd.merge(ScimEn_all, Energy, how='inner', on='Country')
Sci_Energy_GDP_all=pd.merge(Sci_Energy_all, GDP, how='inner', left_on='Country', right_on="Country Name")

#how many entries did you lose by substracting number of rows in Sci_Energy_GDP from Sci_Energy_GDP_all  
len(Sci_Energy_GDP_all)-len(Sci_Energy_GDP)
```

```python
# Step 5



#we will create a new col called avgGDP and its values from the average of gdp vaules from 2006:2015

Sci_Energy_GDP["avgGDP"]=Sci_Energy_GDP[['2006', '2007', '2008',
       '2009', '2010', '2011', '2012', '2013', '2014', '2015']].mean(axis=1)

# create  a Series named avgGDP with 15 countries and their average GDP sorted in descending order.
avgGDP=Sci_Energy_GDP.loc[:,"avgGDP"]

#sorting the series on the values of their avgGDP in a descending order.
avgGDP=avgGDP.sort_values(ascending=False)
display(avgGDP)
type(avgGDP)
```

```python
#step_6


# What is the mean Energy Supply per Capita?


Average_En_supply=Sci_Energy_GDP[['Energy Supply per capita']].mean(axis=0)
display(Average_En_supply)
```

```python
# Step 7


# What country has the maximum % Renewable and what is the percentage?

#sorting the rows in Sci_Energy_GDP depending on the % Renewable col from max to min
Sci_Energy_GDP = Sci_Energy_GDP.sort_values(by=['% Renewable'], ascending=False)

#getting the first row (max ['% Renewable'] ) 

first_row=Sci_Energy_GDP.iloc[0]

#creating a tuble contains name of country( index) and ["% Renewable"] value
count_val=first_row.name,first_row.loc["% Renewable"]
display(count_val)
type(count_val)
```

```python
# Step 8


# Create a new column that is the ratio of Self-Citations to Total Citations

Sci_Energy_GDP["Citation_Ratio"]=Sci_Energy_GDP["Self-citations"]/Sci_Energy_GDP["Self-citations"].sum(axis=0)

#What is the maximum value for this new column, and what country has the highest ratio?

#sorting the Sci_Energy_GDP depending on values of ['Citation_Ratio'] col
Sci_Energy_GDP = Sci_Energy_GDP.sort_values(by=['Citation_Ratio'], ascending=False)

# getting the first row = the max ["Citation_Ratio"]
first_row_2=Sci_Energy_GDP.iloc[0]

#creating a tuble for the country and its val of ["Citation_Ratio"]
count_val_2=first_row.name,first_row_2.loc["Citation_Ratio"]
display(count_val_2)
```

```python
# Step 9

# Create a  column ['flag']  with a 1 if the country's % Renewable value is at or
# above the median for all countries in the top 15, and a 0 if the country's % Renewable value is below the median.

Sci_Energy_GDP['flag'] = np.where(Sci_Energy_GDP['% Renewable']>=Sci_Energy_GDP['% Renewable'].median(axis=0), 1, 0)


# sorting the rows depending on the values of ['Rank'] col 
Sci_Energy_GDP = Sci_Energy_GDP.sort_values(by=['Rank'], ascending=True)

#  series named HighRenew whose index is the country name sorted in ascending order of rank.
HighRenew=Sci_Energy_GDP.loc[:,"flag"]
display(HighRenew)
```

```python
# Step 10

# Use the following dictionary to group the Countries by Continent,
ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}


#converting the dictionar into a dateframe that displays the sample size
Continent = pd.DataFrame.from_dict(ContinentDict,orient ='index')
Continent.reset_index(drop=False, inplace=True)
Continent= Continent.set_axis(['Country','Continent'], axis=1, inplace=False)


# #preparing Sci_Energy_GDP by creating a new col ["Population"] 
Sci_Energy_GDP["Population"]=Sci_Energy_GDP["Energy Supply"]/Sci_Energy_GDP["Energy Supply per capita"]
Sci_Energy_GDP.reset_index(drop=False, inplace=True)


# #merging both dataframes by the country col
cont_count_pop=pd.merge(Continent, Sci_Energy_GDP[["Country","Population"]], how='inner', on='Country')



# #using group by and agg to show values of [size]]number of countries in each continent 
# and  the sum, mean, and std deviation for the estimated population of each country.
req_df=cont_count_pop.groupby("Continent")["Population"].agg([np.size, np.sum, np.mean, np.std]).reset_index()
req_df=req_df.set_index("Continent")
display(req_df)


```
